```python
import numpy as np
import pandas as pd
import plotly_express as px
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
data = pd.read_clipboard(sep='\t',index_col=['week'])
```

```python
data2 = pd.DataFrame(columns = ['week','sjfd','t_week'])
for i in data.columns:
    a = pd.DataFrame(data[i])
    a['t_week'] = i
    a.reset_index(inplace=True)
    a.columns = ['week','sjfd','t_week']
    data2 = pd.concat([data2,a])
```

```python
data2.head()
```

```python
px.scatter(data2,x='week',y='sjfd',color="t_week")
```

```python
px.line(data2,x='week',y='sjfd',color="t_week",line_group="t_week")
```

```python
data2.plot()
```

```python
data5 = data2.copy()
```

```python
data5 = data5.dropna(axis=0,how='any')
```

```python
data5 = data5[data5['week']!=0]
```

```python
data5.head()
```

```python
xdata = np.array(data5['week'])
```

```python
ydata = np.array(data5['sjfd'])
```

```python
def func(x, a, b):
    return a * np.power(x,b)
```

```python
popt, pcov = curve_fit(func, xdata, ydata)
```

```python
y2 = [func(i, popt[0],popt[1]) for i in xdata]
```

```python
plt.plot(xdata,y2,ydata,'r--')
```

```python
print(popt)
```
